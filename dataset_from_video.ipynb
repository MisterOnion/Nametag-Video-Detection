# Extracting Image Dataset from Video

Source:
https://www.kaggle.com/code/amolambkar/create-your-own-image-dataset-using-video

```python
import cv2
import os

video_path = r"C:\Users\limsi\OneDrive\Desktop\NameTag Identification Video Project\sample.mp4"
video = cv2.VideoCapture(video_path)

success = True
count = 1
image_id = 1
output_path = r"C:\Users\limsi\OneDrive\Desktop\NameTag Identification Video Project\extracted_frames"


while success:
    success, frame = video.read()

    # get every 3th frame from video (usually every 5th, but this video is short)
    # no need al frames from video, unless decided otherwise
    if success == True:
        if count%3 == 0:
            # specify output path and file name 
            name = str(image_id) + ".jpg"
            image_id += 1

            os.makedirs(output_path, exist_ok = True)
            save_output_folder = os.path.join(output_path, name)
            # write saved frames to folder
            cv2.imwrite(save_output_folder, frame)
        count += 1
    else:
        break

print("Total Extracted Frames :", image_id)

```
